# Define Monte Carlo Function
Define the monte_carlo_pi function to estimate the value of π using the Monte Carlo method.

```python
def monte_carlo_pi(num_samples):
    inside_circle = 0

    for _ in range(num_samples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1

    return (inside_circle / num_samples) * 4
```

```python
import numpy as np

def newton_method(f, f_prime, initial_guess, tolerance=1e-10, max_iterations=10):
    x = initial_guess
    xplot = []
    for _ in range(max_iterations):
        xplot.append(x)
        x_new = x - f(x) / f_prime(x)
        if abs(x_new - x) < tolerance:
            print(xplot)
            return x_new, xplot
            
        x = x_new
    raise ValueError("Newton's method did not converge")

# Define the function and its derivative
f = np.sin
f_prime = np.cos

# Initial guess
initial_guess = 2

# Estimate π using Newton's method
pi_estimate_newton, xplot = newton_method(f, f_prime, initial_guess)
print(f"Estimated value of π using Newton's method: {pi_estimate_newton}")

# Plot the Estimates for Pi using Newton's Method
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(xplot) + 1), xplot, marker='o', linestyle='-', color='b', label='Newton Method')
plt.axhline(y=np.pi, color='r', linestyle='-', label='Actual π')

# Calculate and plot the log of the error to π
log_errors = [np.log(abs(np.pi - x)) for x in xplot]
print(log_errors)
MethodB = log_errors
plt.plot(range(1, len(log_errors) + 1), log_errors, marker='x', linestyle='--', color='g', label='Log Error to π')

plt.xlabel('Number of Iterations')
plt.ylabel('Estimated Value of π / Log Error')
plt.title('Estimation of π using Newton\'s Method')
plt.legend()
plt.show()
plt.plot(range(1, len(xplot) + 1), xplot, marker='o', linestyle='-', color='b', label='Newton Method')
plt.axhline(y=np.pi, color='r', linestyle='-', label='Actual π')
plt.xlabel('Number of Iterations')
plt.ylabel('Estimated Value of π')
plt.title('Estimation of π using Newton\'s Method')
plt.legend()
plt.show()
```

# Estimate π with a new methodology


```python
def nilakantha_series(iterations):
    pi_estimate = 3.0
    sign = 1
    pi_values = [pi_estimate]
    for i in range(1, iterations + 1):
        term = sign * (4.0 / ((2 * i) * (2 * i + 1) * (2 * i + 2)))
        pi_estimate += term
        pi_values.append(pi_estimate)
        sign *= -1
    return pi_estimate, pi_values

# Estimate π using the Nilakantha series with 1000 iterations
iterations = 20
pi_estimate_nilakantha, pi_values_nilakantha = nilakantha_series(iterations)
print(f"Estimated value of π using Nilakantha series: {pi_estimate_nilakantha}")

# Plot the Estimates for Pi using Nilakantha Series
plt.figure(figsize=(10, 6))
plt.plot(range(iterations + 1), pi_values_nilakantha, marker='o', linestyle='-', color='b', label='Nilakantha Series')
plt.axhline(y=np.pi, color='r', linestyle='-', label='Actual π')
plt.xlabel('Number of Iterations')
plt.ylabel('Estimated Value of π')
plt.title('Estimation of π using Nilakantha Series')
plt.legend()
plt.show()
# Calculate and plot the log of the error to π for Nilakantha Series
log_errors_nilakantha = [np.log(abs(np.pi - x)) for x in pi_values_nilakantha]
MethodC = log_errors_nilakantha

plt.figure(figsize=(10, 6))
plt.plot(range(iterations + 1), log_errors_nilakantha, marker='x', linestyle='--', color='g', label='Log Error to π (Nilakantha Series)')
plt.xlabel('Number of Iterations')
plt.ylabel('Log Error')
plt.title('Log Error to π using Nilakantha Series')
plt.legend()
plt.show()
```

```python
# plot_errors.py
import matplotlib.pyplot as plt
import numpy as np

def plot_log_errors(errors_dict):
    """
    Plots log errors from different methods on the same graph.

    Parameters:
    errors_dict (dict): A dictionary where keys are method names and values are lists of errors.
    """
    plt.figure(figsize=(10, 6))

    for method, errors in errors_dict.items():
        log_errors = np.log(errors)
        plt.plot(log_errors, label=method)

    plt.xlabel('Iteration')
    plt.ylabel('Log Error')
    plt.title('Log Errors from Different Methods')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# plot_errors.py
import matplotlib.pyplot as plt
import numpy as np

def plot_log_errors(errors_dict, max_iterations=40):
    """
    Plots log errors from different methods on the same graph up to a specified number of iterations.

    Parameters:
    errors_dict (dict): A dictionary where keys are method names and values are lists of errors.
    max_iterations (int): The maximum number of iterations to plot.
    """
    plt.figure(figsize=(10, 6))

    for method, errors in errors_dict.items():
        log_errors = errors[:max_iterations]
        plt.plot(log_errors, label=method)

    plt.xlabel('Iteration')
    plt.ylabel('Log Error')
    plt.title('Log Errors from Different Methods')
    plt.legend()
    plt.grid(True)
    plt.show()

errors = {
    'Monte Carlo Integration': MethodA,
    'Newton Root Finding': MethodB,
    'Nilakantha Method': MethodC
}
plot_log_errors(errors, max_iterations=40)
```
